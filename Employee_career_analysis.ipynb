2.WHAT ARE THE FACTORS INFLIUENCING CARRIER ASPIRATIONS OF GENZ ?

```python
ca=data['Which of the below factors influence the most about your career aspirations ?'].value_counts()
label=ca.index
count=ca.values
color=['blue','magenta','aqua','gold','grey']
plt.pie(ca,labels=label,autopct='%0.1f%%',colors=color,startangle=90,radius=1.1,textprops={'fontsize':15},
counterclock=False,labeldistance=1.1,wedgeprops={'linewidth':2,'edgecolor':'black'})
plt.title('factor influencing career aspirations')
plt.legend(loc='center left',bbox_to_anchor=(1,.5, 0.5, 1))
plt.show()
```

3.HOW MANY WANT TO PURSUE HIGHER EDUCATION OUTSIDE INDIA WITH THEIR INVESTMENT?

```python
q2=data['Would you definitely pursue a Higher Education / Post Graduation outside of India ? If only you have to self sponsor it.'].value_counts()
label=q2.index
count=q2.values
color=['blue','green','yellow']
plt.pie(q2,labels=label,colors=color,autopct='%0.1f%%',radius=1,wedgeprops={'linewidth':2,'edgecolor':'black'},textprops={'fontsize':10})
plt.title('Higher education/post graduation place')
plt.legend(loc='center left',bbox_to_anchor=(1,0.5,1,.5))
plt.show()
```

4.HOW LIKELY GENZ IS TO WORK FOR ONE COMPANY FOR THREE YEARS OR MORE ?

```python
q3=data['How likely is that you will work for one employer for 3 years or more ?'].value_counts()
label=q3.index
count=q3.values
color=['gold','aqua','blue']
plt.pie(q3,labels=label,colors=color,radius=1,wedgeprops={'linewidth':2,'edgecolor':'red'},autopct='%0.1f%%')
plt.title('employer work time interval')
plt.legend(loc='upper right',bbox_to_anchor=(1,0,1,1))
plt.show()
```

WHAT IS THE PREFERRED ENVIRONMENT OF GENZ ?

```python
q4=data['What is the most preferred working environment for you.'].value_counts()
label=q4.index
color=['gold','aqua','magenta','blue','green','orange']
plt.pie(q4,labels=label,colors=color,wedgeprops={'linewidth':2,'edgecolor':'black'},autopct='%0.1f%%')
plt.legend(loc='center right',bbox_to_anchor=(1,0.8,1,1))
plt.title("work environment")
plt.show()
```
